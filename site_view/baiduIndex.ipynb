```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains
import time
from IPython.display import Image
```

```python
chrome_options = Options()
# 无头模式启动
chrome_options.add_argument('--headless')
# 谷歌文档提到需要加上这个属性来规避bug
chrome_options.add_argument('--disable-gpu')

chrome_options.add_argument("--window-size=1960,1080");
# 初始化实例
driver= webdriver.Chrome('/opt/chrome/chromedriver',chrome_options=chrome_options)
```

### 访问首页

```python
driver.get("http://index.baidu.com/#/")
```

```python
driver.save_screenshot('screenshot/index.png')
Image('screenshot/index.png')
```

```python
login = driver.find_element_by_class_name('username-text')
print(login.text)
driver.save_screenshot('screenshot/login-panel.png')
Image('screenshot/login-panel.png')

if login.text.strip() == '登录':
    print('start login')
    login.click()
    time.sleep(1)
    input_username = driver.find_element_by_id('TANGRAM__PSP_4__userName')
    print(input_username.text)
    input_username.send_keys('18001053507')
    input_password = driver.find_element_by_id('TANGRAM__PSP_4__password')
    print(input_password.text)
    input_password.send_keys('dib20180606')
    driver.find_element_by_id('TANGRAM__PSP_4__submit').click()
    time.sleep(1)
    driver.save_screenshot('screenshot/input.png')
    Image("screenshot/input.png")
print('end')
```

```python
search_index_list = []
size = canvas[0].size
width = size['width']
for i in range(30):
    ActionChains(driver).move_to_element_with_offset(canvas[0],width-width/29*i-width/60,200).perform()
    #time.sleep(0.1)
    date = driver.find_elements_by_xpath("//div[contains(@style,'9999999')]/div[1]")[0].text
    label = driver.find_elements_by_xpath("//div[contains(@style,'9999999')]/div[2]/div[1]")[0].text
    index = driver.find_elements_by_xpath("//div[contains(@style,'9999999')]/div[2]/div[2]")[0].text.strip()
    search_index_list.append((date.split(),label,index))
print(search_index_list)        
```

```python
print(canvas)
info_index_list = []
size = canvas[2].size
width = size['width']
for i in range(30):
    ActionChains(driver).move_to_element_with_offset(canvas[2],0+(width-10)/30*i,100).perform()
    date = driver.find_elements_by_xpath("//div[contains(@style,'9999999')]/div[1]")[1].text
    label = driver.find_elements_by_xpath("//div[contains(@style,'9999999')]/div[2]/div[1]")[1].text
    index = driver.find_elements_by_xpath("//div[contains(@style,'9999999')]/div[2]/div[2]")[1].text.strip()
    info_index_list.append((date,label,index))
print(info_index_list) 
```
