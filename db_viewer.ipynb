###  使用DB Viewer来预览抓取的数据

使用spider_db.client的SpiderDBClient来连接数据库，数据库为mongo

```python
from spider_db.client import SpiderDBClient
```

### 统计局数据


```python
tongjiju_client = SpiderDBClient(host='10.0.50.141',user='db_user',passwd='12345678')
```

```python
tongjiju_db=tongjiju_client.use_db('tongjiju')
```

 ##### 查看所有统计局的数据表

```python
tongjiju_db.display_docs_readable(1)
```

 ##### 查看所有统计局有多少数据表

```python
print(tongjiju_db.docs_count())
```

 ##### 查找表含有关键字的表
 * 关键字可以在表的名称（id_name），路径名(trace_name)和区域名（title）中
 * 使用docs_readable可以显示并返回简化的表信息，最后一个字段是表地址元组（title，trace，id），可以用这个tuple找到对应的数据表

```python
keyword = "教育"
```

```python
query_tables = tongjiju_db.find_table_by_keyword(keyword)
tongjiju_db.docs_readable(query_tables)
```

##### 通过地址元组来获取数据表


```python
table = tongjiju_db.get_table(section='AD0F',trace=['AD'],id='AD0O')
```

##### 获取数据表内容

* 因为一个表可以有多个显示选项，所以首先需要获得数据表的所有可选项

```python
options = table.get_options()
print(options)
```

* 表选项的格式为[{选项类型1：{ options:[(选项id1，解释)，（选项id2，解释）]}}]
* 可以通过构建选项参数来获取数据 [(类型,id),(类型,id)]

```python
option = [('reg','110000'),('sj','LAST10')]
```

* 可以通过选项来获取数据

```python
data = table.fetch_data(option)
print(table.to_readable())
```

* 可以构建一个随机选择的选项

```python
option_random = []
for k in options.keys():
    option_random.append((k,random.choice(options[k]['options'])[0]))
print(option_random)
```

```python
data = table.fetch_data(option_random)
#print(table.to_readable())
#print(json.dumps(data['htmltable'], indent=2))
display(HTML(data['htmltable']))
```
